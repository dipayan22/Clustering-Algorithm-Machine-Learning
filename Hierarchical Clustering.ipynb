```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import datasets
%matplotlib inline
```

```python
iris=datasets.load_iris()
```

```python
iris_df=pd.DataFrame(iris.data)
```

```python
iris_df
```

```python
iris_df.columns=iris.feature_names
```

```python
iris_df.head()
```

```python
# standardization
from sklearn.preprocessing import StandardScaler

scaler=StandardScaler()

```

```python
X_scaled=scaler.fit_transform(iris_df)
```

```python
X_scaled
```

```python
#Apply PCA
from sklearn.decomposition import PCA

pca=PCA(n_components=2)
```

```python
pca
```

```python
pca_scaled=pca.fit_transform(X_scaled)

pca_scaled
```

```python
plt.scatter(pca_scaled[:,0],pca_scaled[:,1])
```

```python
##Aggromerative Clusttering
##to constraut a dendogram

import scipy.cluster.hierarchy as sc

#plot the dendogram
plt.figure(figsize=(20,7))
plt.title("Dendogram")

##create dendogram
sc.dendrogram(sc.linkage(pca_scaled,method='ward'))
plt.title('Dendogram')
plt.xlabel('Sample Index')
plt.ylabel('Eucledian Distance')
```

```python
from sklearn.cluster import AgglomerativeClustering

cluster=AgglomerativeClustering(n_clusters=2,affinity='euclidean',linkage='ward')
cluster.fit(pca_scaled)
```

```python
cluster.labels_
```

```python
plt.scatter(pca_scaled[:,0],pca_scaled[:,1],c=cluster.labels_)
```

```python
from sklearn.metrics import silhouette_score
```

```python
silhouetter_coefficients=[]

for k in range(2,11):
  agglo=AgglomerativeClustering(n_clusters=k,affinity='euclidean',linkage='ward')
  agglo.fit(X_scaled)
  score=silhouette_score(X_scaled,agglo.labels_)
  silhouetter_coefficients.append(score)
```

```python
plt.plot(range(2,11),silhouetter_coefficients)
plt.xticks(range(2,11))
plt.xlabel("Number of Cluster")
plt.ylabel("Silhouetter Coefficient")
plt.show()
```
